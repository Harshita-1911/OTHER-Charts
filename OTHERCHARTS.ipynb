```python
#a. Display Distributions of Continuous Variables:
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the dataset
df = pd.read_csv("heart_disease_health_indicators_BRFSS2015 2 (1).csv")

# Define the continuous columns to visualize
continuous_columns = ['BMI', 'MentHlth', 'PhysHlth']

# Set up the matplotlib figure
fig, axes = plt.subplots(2, 3, figsize=(18, 10))

# Plot histograms for each continuous variable
for ax, col in zip(axes[0], continuous_columns):
    sns.histplot(df[col], kde=True, ax=ax, color='skyblue')
    ax.set_title(f'Distribution of {col}')
    ax.set_xlabel(col)
    ax.set_ylabel('Frequency')

# Plot box plots for each continuous variable
for ax, col in zip(axes[1], continuous_columns):
    sns.boxplot(x=df[col], ax=ax, color='lightgreen')
    ax.set_title(f'Box Plot of {col}')
    ax.set_xlabel(col)
    ax.set_ylabel('Value')

# Adjust layout
plt.tight_layout()
plt.show()
```

```python
#b. Heatmaps and Correlation Plots: Explore Variable Relationships
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the dataset
df = pd.read_csv("heart_disease_health_indicators_BRFSS2015 2 (1).csv")

# Compute the correlation matrix
correlation_matrix = df.corr()

# Set up the matplotlib figure
plt.figure(figsize=(16, 12))

# Draw the heatmap
sns.heatmap(correlation_matrix, annot=True, fmt=".2f", cmap='coolwarm', linewidths=0.5)

# Set the title
plt.title('Correlation Matrix Heatmap')

# Show the plot
plt.show()
```

```python
#d. Line Plots: Display trends over time or across ordered categories.
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the dataset
df = pd.read_csv("heart_disease_health_indicators_BRFSS2015 2 (1).csv")

# Create age group labels
age_labels = ["18-24", "25-29", "30-34", "35-39", "40-44", "45-49", "50-54", "55-59", "60-64", "65-69", "70-74", "75-79", "80+"]
df['AgeGroup'] = pd.cut(df['Age'], bins=len(age_labels), labels=age_labels, right=False)

# Calculate heart disease prevalence by age group
age_group_heart_disease = df.groupby('AgeGroup')['HeartDiseaseorAttack'].mean()

# Calculate heart disease prevalence by income level
income_heart_disease = df.groupby('Income')['HeartDiseaseorAttack'].mean()

# Set up the matplotlib figure
fig, axes = plt.subplots(1, 2, figsize=(14, 6))

# Line plot: Heart disease prevalence by age group
sns.lineplot(x=age_group_heart_disease.index, y=age_group_heart_disease.values, ax=axes[0], marker='o')
axes[0].set_title('Heart Disease Prevalence by Age Group')
axes[0].set_xlabel('Age Group')
axes[0].set_ylabel('Prevalence of Heart Disease')

# Line plot: Heart disease prevalence by income level
sns.lineplot(x=income_heart_disease.index, y=income_heart_disease.values, ax=axes[1], marker='o')
axes[1].set_title('Heart Disease Prevalence by Income Level')
axes[1].set_xlabel('Income Level')
axes[1].set_ylabel('Prevalence of Heart Disease')

# Adjust layout
plt.tight_layout()
plt.show()
```

```python
#e. Violin Plots: Combine aspects of box plots and density plots to show the distribution of data.
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the dataset
df = pd.read_csv("heart_disease_health_indicators_BRFSS2015 2 (1).csv")

# Set up the matplotlib figure
fig, axes = plt.subplots(1, 2, figsize=(14, 6))

# Violin plot: Distribution of BMI by Heart Disease
sns.violinplot(x='HeartDiseaseorAttack', y='BMI', data=df, ax=axes[0], palette='coolwarm')
axes[0].set_title('Distribution of BMI by Heart Disease')
axes[0].set_xlabel('Heart Disease (0: No, 1: Yes)')
axes[0].set_ylabel('BMI')

# Violin plot: Distribution of Mental Health Days by Heart Disease
sns.violinplot(x='HeartDiseaseorAttack', y='MentHlth', data=df, ax=axes[1], palette='coolwarm')
axes[1].set_title('Distribution of Mental Health Days by Heart Disease')
axes[1].set_xlabel('Heart Disease (0: No, 1: Yes)')
axes[1].set_ylabel('Mental Health Days')

# Adjust layout
plt.tight_layout()
plt.show()
```

```python
#f. Pair Plots: Plot pairwise relationships across multiple variables.
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load the dataset
df = pd.read_csv("heart_disease_health_indicators_BRFSS2015 2 (1).csv")

# Select a subset of variables to plot
variables = ['BMI', 'MentHlth', 'PhysHlth', 'Age', 'HeartDiseaseorAttack']

# Create a pair plot
sns.pairplot(df[variables], hue='HeartDiseaseorAttack', palette='coolwarm')

# Display the plot
plt.show()
```
